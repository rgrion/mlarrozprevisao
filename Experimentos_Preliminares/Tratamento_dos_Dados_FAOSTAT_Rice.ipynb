```python
df = pd.read_csv('FAOSTAT_Full_Rice_2023.csv')
df
```

```python
df.shape
```

```python
df_reduzido = df[['Area', 'Year', 'Element', 'Value']]

df_pivot = df_reduzido.pivot_table(index=['Area', 'Year'],
                                   columns='Element',
                                   values='Value').reset_index()

df_pivot.rename(columns={
    'Area harvested': 'Area harvested (ha)',
    'Yield': 'Yield (kg/ha)',
    'Production': 'Production (t)'
}, inplace=True)

df_pivot
```

```python
df_pivot.shape
```

```python
df_pivot.to_csv('/content/Rice_2011.csv', index=False)
```
